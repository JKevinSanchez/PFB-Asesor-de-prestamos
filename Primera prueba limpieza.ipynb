```python
n_nans = df.isnull().sum().to_frame(name = "null_counts")
n_nans.reset_index(inplace=True)
n_nans.columns = ["columna", "numero_nans"]
n_nans
```

```python
pd.set_option("display.max_rows", None)
n_nans
```

```python
# Filtramos el dataframe original para eliminar columnas con una gran cantidad de nans
numero = 500000 # Podemos cambiar el numero a mas o menos nans

numero_nans = df.isnull().sum()

columnas_menos_nans = numero_nans[numero_nans < numero].index

df_filtrado = df[columnas_menos_nans]
df_filtrado.head()
```

```python
pd.set_option("display.max_rows", 30)

df_filtrado
```

```python
# Analizando el dataframe e investigando, vamos a eliminar ciertas columnas irrelevantes

columnas_irrelevantes = [
    "id",
    "url",
    "emp_title",
    "title",
    "zip_code",
    "last_pymnt_d",
    "last_pymnt_amnt",
    "last_credit_pull_d",
    "disbursement_method",
    "policy_code",
    "hardship_flag"
]

df_filtrado = df_filtrado.drop(columns = columnas_irrelevantes)
df_filtrado
```

```python
# Visualizaciones para seguir analizando para limpieza e importancia

import seaborn as sns
import matplotlib.pyplot as plt

df_numerico = df_filtrado.select_dtypes(include = ["number"])
correlacion = df_numerico.corr()

plt.figure(figsize=(20, 15))
sns.heatmap(
    correlacion,
    annot = True,
    cmap = "coolwarm"
    )
plt.show()
```
